First we will load the python library and connect to the data. 

```python
import pandas as pd
import matplotlib.pyplot as plt
filename = 'DailyChangeinCasesbyPHU.csv'
daily_covid_cases = pd.read_csv(filename)
daily_covid_cases.head()
```

## Descriptive Statistics

Visualizing the dataset using simple descriptive statistics and summary measures.

```python
# Determine the number of days included in the dataset
len(daily_covid_cases)
```

```python
# Examine the dataset types
daily_covid_cases.info()
```

```python
# Checking summary statistics for the dataframe
daily_covid_cases.describe()
```

In the table above, we can use the mean number of cases to approximate which Public Health units managed the most COVID-19 cases (Toronto). Future analysis should calculate rates to allow for inclusion of population size to make judgements about Public Health jurisdictions most impacted.

```python
# Identify the dates with the highest and lowest case counts
daily_covid_cases['Total'].min()
```

```python
daily_covid_cases['Total'].max()
```

```python
# Check the yearly cases across the province

# Ensure 'Date' is in datetime format
daily_covid_cases['Date'] = pd.to_datetime(daily_covid_cases['Date'])

# Extract the year correctly
daily_covid_cases['Year'] = daily_covid_cases['Date'].dt.year  # Use .dt.year

# Group by year and sum 'Total' column
yearly_summary = daily_covid_cases.groupby('Year')['Total'].sum().reset_index()

# Rename columns for clarity
yearly_summary.columns = ['Year', 'Total Cases']

# Print the summary table
print(yearly_summary)
```

## Visualization
Now that we've completed some simple descriptive analysis, lets now visualize the provincial cases across time. 

```python
# Ensure 'date' is a datetime column
daily_covid_cases['Date'] = pd.to_datetime(daily_covid_cases['Date'])

# Sort by date in case it's not sorted
daily_covid_cases = daily_covid_cases.sort_values('Date')

# Plot total cases over time
plt.figure(figsize = (12, 6))
plt.plot(daily_covid_cases['Date'], daily_covid_cases['Total'], label='Daily Cases', linewidth=1)

# Labels and title
plt.xlabel('Date')
plt.ylabel('Total COVID-19 Cases')
plt.title('Daily COVID-19 Cases in Ontario (2020-2024)')
plt.legend()
plt.grid(True)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
